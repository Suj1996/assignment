```python
#3 Explain the following terms:
* the difference in np.array,np.asarray and np.asanarray.
* the difference between deep copy and shallow from array.
```

```python
np.array, np.asarray, and np.asanyarray are all functions provided by the NumPy library in Python for creating arrays. While they seem similar, there are subtle differences between them:

#np.array:
np.array is a function that creates a new NumPy array from an existing array-like object (such as lists or tuples) or from a Python iterable (like range).
If the input is already a NumPy array, np.array will create a new copy of the array by default. However, you can specify the copy=False argument to avoid copying if the input is already a NumPy array.
This function generally behaves like a constructor for creating new arrays.

#np.asarray:
np.asarray is a function that performs a similar function to np.array. It converts the input to an array, but if the input is already a NumPy array, it does not make a copy by default. Instead, it returns the original array.
If the input is a subclass of ndarray, such as a matrix, then np.asarray will convert it to a NumPy array. Otherwise, it will create a new array.
np.asarray is particularly useful when you want to ensure that an input is converted to a NumPy array, but you don't want to unnecessarily copy the data if it's already in the correct format.

#np.asanyarray:
np.asanyarray is a function that also converts the input to an array. However, it differs from np.asarray in that it will convert input types that are not subclasses of ndarray into arrays, whereas np.asarray will convert them into ndarray subclasses if possible.
If the input is already an ndarray, np.asanyarray does not make a copy by default, similar to np.asarray.
The purpose of np.asanyarray is to create a NumPy array if the input is not already an array, but to return the input array if it is already an array, regardless of its subclass.

#In summary:
Use np.array when you want to create a new array from an existing array-like object or Python iterable.
Use np.asarray when you want to ensure that an input is converted to a NumPy array, but you don't want to make a copy if the input is already an array.
Use np.asanyarray when you want to ensure that an input is converted to a NumPy array, but you want to retain the original array's subclass if it's already an ndarray, or convert other types to ndarray.

#Shallow Copy:
A shallow copy creates a new object, but it doesn't create copies of nested objects within the original object. Instead, it copies references to the nested objects. So, if you modify a nested object in the shallow copy, it will affect the original object and vice versa.
Shallow copies are usually faster and use less memory than deep copies because they don't duplicate the nested objects.
In Python, you can create a shallow copy using the copy() method for mutable objects like lists, dictionaries, or sets. Additionally, the copy module provides the copy() function to create shallow copies.
Example: If you have a list containing other lists, a shallow copy of the outer list will create a new list object, but it will still reference the same inner lists as the original.
    
#Deep Copy:
A deep copy, on the other hand, creates a completely new object with completely new copies of all nested objects within the original object. So, modifying a nested object in a deep copy won't affect the original object, and vice versa.
Deep copies ensure that changes made in one object do not affect the other, making them safer when working with mutable nested objects.
In Python, you can create a deep copy using the copy.deepcopy() function from the copy module. This function recursively copies all nested objects.
Deep copies can be slower and consume more memory than shallow copies because they create entirely new copies of nested objects.
Example: If you have a list containing other lists, a deep copy of the outer list will create a new list object and recursively create deep copies of all the inner lists as well.

```

```python
#14 Create a time series dataset in a pandas datafreame with column :date,temprature,humidity and perform the following task using matplotlib and numpy:
a)plot the temperature and humidity on the same plot with diferent y-axes (left y-axis for temperature and right y-axis for humidity).
b)label the x-axis as Date.
c)set the title the plot as Temperature and humidity over time.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
date_range = pd.date_range(start='2024-01-01', end='2024-01-31', freq='D')
temperature = np.random.randint(0, 35, size=len(date_range))
humidity = np.random.randint(20, 100, size=len(date_range))
data = {'Date': date_range, 'Temperature': temperature, 'Humidity': humidity}
df = pd.DataFrame(data)
fig, ax1 = plt.subplots()
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature')
ax1.plot(df['Date'], df['Temperature'])
ax1.tick_params(axis='y')
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity')
ax2.plot(df['Date'], df['Humidity'])
ax2.tick_params(axis='y',)
plt.title('Temperature and Humidity over Time')
plt.show()
```
