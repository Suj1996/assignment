```python
#Q5 generate a synthetic dataset with correlated features.visualize tge correlation matrix of a dataset using a heatmap.
```

```python
# Generate correlated data
np.random.seed(10)
data = np.random.multivariate_normal(mean=[0, 0], cov=[[1, 0.75], [0.75, 1]], size=100)
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2'])
# Calculate correlation matrix
correlation_matrix = df.corr()
# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix)

# Add labels and title
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=35)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation Matrix')

# Show heatmap
plt.tight_layout()
plt.show()
```

```python
#3 Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the DataFrame to get a matrix format for the heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap
fig = go.Figure(data=go.Heatmap(
    z=pivot_table.values,
    x=pivot_table.columns,
    y=pivot_table.index,
    colorscale='YlGnBu'
))

# Update layout for better readability
fig.update_layout(
    title='Heatmap of Sales Variation Across Months and Days',
    xaxis_title='Month',
    yaxis_title='Day',
    yaxis=dict(dtick=1)  # Ensure all days are shown
)

fig.show()
```

```python
#Q4. Using the given x and y data, generate a 3D surface plot to visualize the function
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
   x, y = np.meshgrid(x, y)
   z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Update layout for better readability
fig.update_layout(
    title='3D Surface Plot of the Function',
    scene = dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show the plot
fig.show()
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Display Bokeh plots in the notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
num_points = 100
x = np.random.random(size=num_points) * 100
y = np.random.random(size=num_points) * 100
sizes = np.random.random(size=num_points) * 30 + 10  # Marker sizes between 10 and 40
colors = np.random.random(size=num_points) * 255  # Colors as grayscale values

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a ColumnDataSource from the DataFrame
source = ColumnDataSource(data)

# Create a new plot
p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add a scatter renderer with legend and different sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=source, legend_label="Random Points")

# Show the results
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Step 1: Import necessary libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Step 2: Display Bokeh plots in the notebook
output_notebook()

# Step 4: Create a ColumnDataSource
# This is a way to store data in a format that Bokeh can easily use
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Step 5: Create a new plot
# x_range specifies the categories on the x-axis (the fruits)
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts')

# Step 6: Add a bar renderer with different colors
# The vbar method is used to create vertical bars in the plot
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white')

# Step 7: Customize the plot
# Remove the grid lines on the x-axis and start the y-axis from 0
p.xgrid.grid_line_color = None
p.y_range.start = 0
# Place the legend at the top center and make it horizontal
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the results
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Step 1: Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Step 2: Display Bokeh plots in the notebook
output_notebook()
# Step 4: Create a new plot
p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency')

# Step 5: Add a quad renderer with the histogram data
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

# Step 6: Customize the plot
p.grid.grid_line_color="white"

# Show the results
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Step 1: Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper

# Step 2: Display Bokeh plots in the notebook
output_notebook()

# Step 4: Create a new plot
p = figure(title="Heatmap", x_axis_label='X-axis', y_axis_label='Y-axis',
           x_range=(0, 1), y_range=(0, 1))

# Step 5: Add the heatmap to the plot
# Use image method to add heatmap
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0, 0), title='Scale')
p.add_layout(color_bar, 'right')

# Show the results
show(p)
```
